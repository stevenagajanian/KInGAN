```python
from bs4 import BeautifulSoup
import pycurl
from io import BytesIO
import pandas as pd
import os.path
```

```python
genes = pd.read_csv('~/Documents/School/Grad_School/Thesis/gan/kinase_genes.csv')
```

```python
gene_names = genes['name']
```

```python
mol2_dict = {}
for i in gene_names.values:
    try:
        b_obj = BytesIO() 
        crl = pycurl.Curl() 

        # Set URL value
        crl.setopt(crl.URL, 'https://zinc15.docking.org/genes/'+i+'/substances.mol2?count=all')

        # Write bytes that are utf-8 encoded
        crl.setopt(crl.WRITEDATA, b_obj)

        # Perform a file transfer 
        crl.perform() 

        # End curl session
        crl.close()
        completeName = os.path.join('/Users/stevenagajanian/Documents/School/Grad_School/Thesis/gan/kinases', i+".mol2") 
        # Get the content stored in the BytesIO object (in byte characters) 
        body = b_obj.getvalue()
        with open(completeName,'w') as file:
            file.write(body)
    except:
        print(i)
```
